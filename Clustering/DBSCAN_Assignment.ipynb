```python
# Importing necesary packages
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
# importing the datset
c_data = pd.read_csv("crime_data.csv")
c_data.head()
```

```python
c_data.info()
```

```python
# removing first column because it's just an name identifier
data = c_data.iloc[:,1:5]
data.head()
```

```python
# checking for null values 
std_df = StandardScaler().fit_transform(data)
std_df
```

# Model Building or Creating a clusters

```python
dbscan = DBSCAN(eps = 1, min_samples = 4)
dbscan.fit(std_df)
```

```python
dbscan.labels_
```

```python
# assigning clusterid to data set
c_data['cluser_id'] = dbscan.labels_
```

```python
c_data.head()
```

```python
# clusters value counts
c_data['cluser_id'].value_counts()
```

```python
# grouping clusters using clustersid
c_data.groupby('cluser_id').agg(['mean'])
```

```python
c_data[c_data.cluser_id == -1]
```

# Airlines 

```python
airlines = pd.read_csv("EastWestAirlines.csv")
airlines.head()
```

```python
# removing the first column which is only a identifier
airlines2 = airlines.iloc[:,1:]
airlines2.head()
```

```python
# cheking for nul-VALUES AND column types
airlines2.info()
```

```python
# standardizing the data 
airlines_norm = StandardScaler().fit_transform(airlines2)
airlines_norm
```

```python
# Creating clusters
airlines_c = DBSCAN(eps = 1, min_samples = 6)
airlines_c.fit(airlines_norm)
```

```python
print(airlines_c.labels_)
```

```python
# assigning clustersid to dataset
airlines2['cluser_id'] = airlines_c.labels_
airlines2
```

```python
airlines2['cluser_id'].value_counts()
```

```python
# grouping clusters using clustersid
airlines2.groupby('cluser_id').agg(['mean'])
```
