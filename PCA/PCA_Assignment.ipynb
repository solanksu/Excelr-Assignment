```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("wine.csv")
```

```python
df.head()
```

```python
df1 = df.iloc[:,1:]
```

```python
df1.head()
```

```python
# converting to array format
W_data = df1.values
```

```python
W_data
```

```python
# normalizing the data 
wine_normal = scale(W_data)
```

```python
wine_normal
```

```python
# instantiatiang an objeact for storing algorithm
pca = PCA()
pca_values = pca.fit_transform(wine_normal)
```

```python
pca_values
```

```python
# pca components
pca.components_
```

```python
# amount of vaience that each pca explains
var = pca.explained_variance_ratio_
var
```

```python
# cumulative varience
var1 = np.cumsum(np.round(var,decimals = 4)*100)
var1
```

```python
# varience plot for pca_components obtained
plt.plot(var1,color="red")
```

```python
pca_values[:,0:3]
```

```python
pca_values[:,1:2]
```

```python
pca_values[:,2:3]
```

```python
x = pca_values[:,0:1]
y = pca_values[:,1:2]
z = pca_values[:,2:3]
```

```python
final_df = pd.concat([df.iloc[:,0]],axis = 1)
```

```python
final_df
```

```python
final_df['pca_1'] = x
final_df['pca_2'] = y  
final_df['pca_3'] = z
```

```python
final_df
```

```python
# visuaization of pca in 2d
fig = plt.figure(figsize = (15,11))
sns.scatterplot(data = final_df)
```

# KMeans

```python
from sklearn.cluster import KMeans
```

### we have already normalized the data above using scale
### we will use the elbow method to find the clusters range

```python
wcss = []
for i in range(1,6):
    kmeans = KMeans(n_clusters = i, random_state = 2)
    kmeans.fit(wine_normal)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1,6), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 3, random_state = 30)
kmeans.fit(wine_normal)
```

```python
kmeans.labels_
```

```python
wine1 = df.copy()
wine1['cluster_id'] = kmeans.labels_ 
```

```python
wine1.head()
```

```python
wine1['cluster_id'].value_counts()
```

# Hierachial_Clustering

```python
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch
```

```python
fig = plt.figure(figsize=(20,11))
dendrogram = sch.dendrogram(sch.linkage(wine_normal,method = 'complete'))
```

```python
hc = AgglomerativeClustering(n_clusters = 3,affinity = 'euclidean', linkage = 'ward')
y_hc = hc.fit_predict(wine_normal)
```

```python
y_hc
```

```python
wine2 = df.copy()
wine2['clusterid'] = y_hc
```

```python
wine2.head()
```

```python
wine2['clusterid'].value_counts()
```
